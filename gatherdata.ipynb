```python
def find_mergeable_dataframes(df_list):
    mergeable_dfs = {}
    for i in range(len(df_list)):
        for j in range(i+1, len(df_list)):
            common_columns = set(df_list[i].columns).intersection(set(df_list[j].columns))
            common_columns = [col for col in common_columns if df_list[i][col].dtype == df_list[j][col].dtype]
            if 'id' in common_columns:
                common_columns.remove('id')
            if common_columns:
                if i in mergeable_dfs:
                    mergeable_dfs[i].append((j, list(common_columns)))
                else:
                    mergeable_dfs[i] = [(j, list(common_columns))]
    return mergeable_dfs

mergeable_dfs = find_mergeable_dataframes(sno_full)
```
